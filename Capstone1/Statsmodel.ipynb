```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Data/2017.csv')
```

```python
df = df.dropna(subset=['DEP_TIME',
       'DEP_DELAY', 'DEP_DELAY_NEW', 'TAXI_OUT', 'WHEELS_OFF', 'WHEELS_ON',
       'TAXI_IN', 'CRS_ARR_TIME', 'ARR_TIME', 'ARR_DELAY', 'ARR_DELAY_NEW',
       'ACTUAL_ELAPSED_TIME', 'AIR_TIME'])
```

```python
df = df[['MONTH','DAY_OF_MONTH', 'DAY_OF_WEEK', 'OP_CARRIER', 'ORIGIN', 'DEST', 'DEP_TIME',
       'ARR_TIME', 'ARR_DELAY']]
```

```python
df.head()
```

```python
objects = ['OP_CARRIER', 'ORIGIN', 'DEST']
```

```python
plt.scatter(df.MONTH, df.ARR_DELAY)
plt.xlabel("Month of trip")
plt.ylabel("Arrival Delay")
plt.title("Relationship between Month of trip and Arrival delay")
```

```python
m = ols('ARR_DELAY ~ MONTH',df).fit()
print(m.summary())
```

```python
plt.scatter(df.DAY_OF_MONTH , df.ARR_DELAY)
plt.xlabel("Day of trip")
plt.ylabel("Arrival Delay")
plt.title("Relationship between Day of trip and Arrival delay")
```

```python
m2 = ols('ARR_DELAY ~ DAY_OF_MONTH',df).fit()
print(m2.summary())
```

```python
plt.scatter(df.DAY_OF_WEEK , df.ARR_DELAY)
plt.xlabel("Day of trip")
plt.ylabel("Arrival Delay")
plt.title("Relationship between weekday of trip and Arrival delay")
```

```python
m3 = ols('ARR_DELAY ~ DAY_OF_WEEK',df).fit()
print(m3.summary())
```

```python
m4 = ols('ARR_DELAY ~ OP_CARRIER',df).fit()
print(m4.summary())
```

```python
plt.scatter(df.DEP_TIME , df.ARR_DELAY, alpha=0.05)
plt.xlabel("Departure Time")
plt.ylabel("Arrival Delay")
plt.title("Relationship between Departure Time and Arrival delay")
```

```python
m5 = ols('ARR_DELAY ~ DEP_TIME',df).fit()
print(m5.summary())
```

```python
plt.scatter(df.ARR_TIME , df.ARR_DELAY, alpha=0.05)
plt.xlabel("Arrival Time")
plt.ylabel("Arrival Delay")
plt.title("Relationship between Arrival Time and Arrival delay")
```

```python
m6 = ols('ARR_DELAY ~ ARR_TIME',df).fit()
print(m6.summary())
```
